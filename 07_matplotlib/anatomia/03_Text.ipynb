# `Text`: criando textos em um plot
- Assim como linhas e retangulos, podemos adicionar textos em um plot
- Para isso, utilizamos a classe `Text`
- Esse texto pode ser adicionado em qualquer região da figura
- Eles são particularmente interessantes para destacar alguma coisa específica em algum plot

```python
import matplotlib.pyplot as plt
from matplotlib.text import Text
```

- O princípio é basicamente o mesmo
- Criar o texto e adicionar em um `Axes`

```python
txt_1 = Text(text='IA em saúde', x=0.1, y=0.7)
txt_2 = Text(text='Matplotlib', x=0.4, y=0.3)


fig, ax = plt.subplots()
ax._add_text(txt_1)
ax._add_text(txt_2)
```

- **Podemos acessar todos os textos de um `Axes` maneira similar aos anteriores**

```python
ax.texts
```

```python
ax.texts[0]
```

## Propriedades e atributos
- Obviamente você pode alterar diversos atributos/propriedades para deixar o texto da maneira que desejar
- Novamente, como é muito similar, não vamos entrar em tantos detalhes
- Vamos apenas fazer uma série de alterações:

```python
ax.texts[0].set_color('blue') # Alterando a cor
ax.texts[0].set_size(16) # Alterando o tamanho
ax.texts[0].set_alpha(0.8) # Alterando opacidade
ax.texts[0].set_family('monospace') # Alterando a família da fonte
ax.texts[0].set_linespacing(0.2) # Alterando espaçamento
ax.texts[0].set_rotation(45) # Alterando rotação
ax.texts[0].set_style('italic') # Alterando o estilo da fonte
ax.texts[0].set_weight('bold') # Alterando o peso da fonte

ax.texts[0].set_x(0.0) # Alterando a posição em X
ax.texts[0].set_y(0.1) # Alterando a posição em Y

ax.figure
```

- **Também podemos usar essas propriedades no construtor**
    - [Documentação](https://matplotlib.org/stable/api/text_api.html#matplotlib.text.Text)

```python
txt_3 = Text(text="TEST", x=0.8, y=0.8, color="red", fontsize=18)
ax._add_text(txt_3)
fig
```

- Ainda é possível customizar uma caixa em volta do texto usando o método `set_bbox()`
    - [Documentação](https://matplotlib.org/3.1.1/api/text_api.html#matplotlib.text.Text.set_bbox)

```python
ax.texts[2].set_bbox(dict(facecolor='lime', boxstyle='larrow, pad=0.4', alpha=0.7))
fig
```
